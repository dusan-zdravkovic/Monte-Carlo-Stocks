## Monte Carlo Method - Geometric Interpretation

### 1. Square and Circle Setup
In the context of the Monte Carlo simulation for estimating $\pi$:
- **Square**: Imagine a square with a side length of 2 units, centered at the origin (from -1 to 1 on both x and y axes).
- **Circle**: A circle with a radius of 1 unit is inscribed within this square.

### 2. Area Relations
- **Area of the Circle**: The area of the circle is given by $\pi r^2$. For a radius \(r\) of 1, the area is $\pi(1^2) = \pi$.
- **Area of the Square**: The area of the square is given by the square of the side length. Since the side is 2 units, the area is $2^2 = 4$ square units.

### 3. Area Ratio
- **Ratio of Areas**: The ratio of the area of the circle to the area of the square is $\frac{\pi}{4}$. This ratio is crucial in the Monte Carlo method to estimate the value of $\pi$, using random point generation within the square.

```python
import numpy as np

def estimate_pi(num_samples):
    # Generate random points within [-1, 1] x [-1, 1]
    x = np.random.uniform(-1, 1, num_samples)
    y = np.random.uniform(-1, 1, num_samples)
    
    # Calculate whether each point falls within the unit circle
    is_inside = x**2 + y**2 <= 1
    
    # Estimate pi
    pi_estimate = 4 * np.sum(is_inside) / num_samples
    return pi_estimate

# Running the function with 1,000,000 samples
estimated_pi = estimate_pi(1000000)
print(f"Estimated value of Pi: {estimated_pi}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def estimate_pi(num_samples):
    # Generate random points within [-1, 1] x [-1, 1]
    x = np.random.uniform(-1, 1, num_samples)
    y = np.random.uniform(-1, 1, num_samples)
    
    # Calculate whether each point falls within the unit circle
    is_inside = x**2 + y**2 <= 1
    
    # Estimate pi based on the ratio of points inside the circle to the total points
    pi_estimate = 4 * np.sum(is_inside) / num_samples
    
    # Plotting
    plt.figure(figsize=(8, 8))  # Set the figure size for better visibility
    plt.scatter(x[is_inside], y[is_inside], color='green', s=1, label='Inside the circle')
    plt.scatter(x[~is_inside], y[~is_inside], color='red', s=1, label='Outside the circle')
    circle = plt.Circle((0, 0), 1, edgecolor='blue', linewidth=2, facecolor='none')
    plt.gca().add_artist(circle)
    plt.xlim(-1, 1)
    plt.ylim(-1, 1)
    plt.gca().set_aspect('equal', adjustable='box')  # Set equal scaling by changing axis limits
    plt.title(f'Estimation of Pi using Monte Carlo method: Pi ≈ {pi_estimate:.4f}')
    plt.legend()
    plt.show()

    return pi_estimate

# Running the function with a sample size of 10,000 for better visualization
estimated_pi = estimate_pi(10000)
print(f"Estimated value of Pi: {estimated_pi}")
```
